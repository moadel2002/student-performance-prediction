```python
#uploading data
df = pd.read_csv('Student Performance Prediction-Binary.csv')
```

```python
df
```

```python
df.drop(columns=['Student ID'],inplace=True)
```

```python
df
```

```python
# X and Y at 20% of the data (quiz1,assignment1)
x_20 =df.iloc[:, 0:2].values
y_20= df.iloc[:, -1].values
```

```python
x_20
```

```python
y_20
```

```python
#solving the inbalanced data using smote technique
smote=SMOTE(sampling_strategy='minority')
x_sm_1,y_sm_1 = smote.fit_resample(x_20,y_20)
```

```python
#checking that the data is now balanced
unique, counts = np.unique(y_sm_1, return_counts=True)
dict(zip(unique, counts))
```

```python
#splitting data
X_train_20, X_test_20, Y_train_20, Y_test_20 = train_test_split(x_sm_1, y_sm_1, test_size = 0.25, random_state = 0)
```

```python
#feature scaling
sc = StandardScaler()
X_train_20 = sc.fit_transform(X_train_20)
X_test_20 = sc.transform(X_test_20)
```

```python
#svc model
# from sklearn.svm import SVC
# classifier = SVC(kernel = 'rbf', random_state = 0) 
# classifier.fit(X_train_20, Y_train_20) 
```

```python
# fitting data into svc model using grid search
C_range = [0.25,0.5,1]
gamma_range = np.logspace(0.3,0.8)
param_grid = dict(gamma=gamma_range, C=C_range)
cv = 3

grid_20 = GridSearchCV(SVC(kernel = 'rbf', random_state = 0), param_grid=param_grid, cv=cv)
grid_20.fit(X_train_20,Y_train_20)

print(
    "The best parameters are %s with a score of %0.2f"
    % (grid_20.best_params_, grid_20.best_score_)
)
```

```python
#predection at 20% of the data
Y_predict_20 = grid_20.predict(X_test_20)
```

```python
Y_predict_20
```

```python
# Displaying accuracy and confusion matrix (TP,FP,FN,TN)
con_matrx = confusion_matrix(Y_test_20, Y_predict_20)
print(con_matrx)
Accuracy=accuracy_score(Y_test_20, Y_predict_20) *100
print("Accuracy of this model at 20% is: " +str(Accuracy) + " %")
```
